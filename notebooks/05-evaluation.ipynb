```python
#w medycynie czesto recall wazniejszy niz precision, bo lepiej wykryc za duzo anomalii niz przeoczyc powazna chorobe

#importy
import os
import sys
import numpy as np
import matplotlib.pyplot as plt

import torch

from torch.utils.data import DataLoader

from sklearn.metrics import (
    classification_report,
    multilabel_confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    auc
)

PROJECT_PATH = r"C:\Users\cp\PycharmProjects\ECGmodel"

os.chdir(PROJECT_PATH)

SRC_PATH = os.path.join(PROJECT_PATH, "src")

if SRC_PATH not in sys.path:
    sys.path.append(SRC_PATH)

from ecg_dataset import ECGDataset
from model import ECGCNN

#device
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print(device)

#load dataset
test_dataset = ECGDataset(
    "data/processed/X_test_ml.npy",
    "data/processed/y_test_ml.npy"
)

test_loader = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle=False
)

#load model
model = ECGCNN(num_classes=5)

model.load_state_dict(
    torch.load(
        "models/ecg_cnn.pth",
        map_location=device
    )
)

model = model.to(device)

model.eval()

print("model loaded")

#predictions
all_probs = []
all_preds = []
all_targets = []

with torch.no_grad():

    for X_batch, y_batch in test_loader:

        X_batch = X_batch.to(device)

        outputs = model(X_batch)

        probs = torch.sigmoid(outputs)

        preds = (probs > 0.5).float()

        all_probs.append(probs.cpu())
        all_preds.append(preds.cpu())
        all_targets.append(y_batch)

#tensor -> numpy
y_prob = torch.cat(all_probs).numpy()

y_pred = torch.cat(all_preds).numpy()

y_true = torch.cat(all_targets).numpy()

print(y_true.shape)
print(y_pred.shape)

#classification report
class_names = [
    "NORM",
    "MI",
    "STTC",
    "CD",
    "HYP"
]

print(
    classification_report(
        y_true,
        y_pred,
        target_names=class_names,
        zero_division=0
    )
)

#confusion matrix
cms = multilabel_confusion_matrix(
    y_true,
    y_pred
)

#plot loop
for i, cm in enumerate(cms):

    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm
    )

    disp.plot()

    plt.title(
        f"Confusion Matrix - {class_names[i]}"
    )

    plt.show()

#roc curve + auc
plt.figure(figsize=(8,6))

for i in range(len(class_names)):

    fpr, tpr, _ = roc_curve(
        y_true[:, i],
        y_prob[:, i]
    )

    roc_auc = auc(fpr, tpr)

    plt.plot(
        fpr,
        tpr,
        label=f"{class_names[i]} AUC={roc_auc:.2f}"
    )

plt.plot([0,1], [0,1], linestyle="--")

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.title("ROC Curves")

plt.legend()

plt.show()

#fajna analiza mean prediction frequency
print("mean predicted probabilities:")

for i, cls in enumerate(class_names):

    print(
        cls,
        y_prob[:, i].mean()
    )

#testowo zmiana tresholdu z domyslnego 0.5 na 0.3 i sprawdzenie jak to wpływa na classification report
#recall powinien wzrosnac a precision spasc, bo model bedzie klasyfikowal wiecej probek jako pozytywne
#preds = (probs > 0.3).float()
```
