```python
#setup
import os
import sys
import numpy as np
import pandas as pd
import wfdb

PROJECT_PATH = r"C:\Users\cp\PycharmProjects\ECGmodel"
os.chdir(PROJECT_PATH)

SRC_PATH = os.path.join(PROJECT_PATH, "src")

if SRC_PATH not in sys.path:
    sys.path.append(SRC_PATH)

from preprocessing import preprocess_pipeline
from labels import encode_labels
from dataset import prepare_multilabel_dataset

#metadata
RAW_PATH = os.path.join(PROJECT_PATH, "data", "raw", "ptbxl")

df = pd.read_csv(
    os.path.join(RAW_PATH, "ptbxl_database.csv")
)

statements = pd.read_csv(
    os.path.join(RAW_PATH, "scp_statements.csv"),
    index_col=0
)

#helper
def get_record_path(filename_lr):
    return os.path.join(RAW_PATH, filename_lr)

#labels
y = encode_labels(df, statements)

print(y.shape)
print(y[:5])

#preprocessing; na poczatku 500 rekordow
signals = []
labels = []

LIMIT = 5000

#load loop
for i, row in df.iloc[:LIMIT].iterrows():

    path = get_record_path(row["filename_lr"])

    if not os.path.exists(path + ".hea"):
        continue

    record = wfdb.rdrecord(path)

    signal = record.p_signal[:, 0]

    segments = preprocess_pipeline(signal)

    for seg in segments:
        signals.append(seg)
        labels.append(y[i])

#numpy
X = np.array(signals, dtype=np.float32)
y = np.array(labels, dtype=np.float32)

print(X.shape)
print(y.shape)

#reshape do cnn - (num_samples, num_timesteps, num_channels)
X = X[:, np.newaxis, :]

#zapis
prepare_multilabel_dataset(X, y)
```
