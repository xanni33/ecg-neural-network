```python
#policzenie class frequency
import numpy as np
import os
import sys
import torch

PROJECT_PATH = r"C:\Users\cp\PycharmProjects\ECGmodel"

os.chdir(PROJECT_PATH)

SRC_PATH = os.path.join(PROJECT_PATH, "src")

if SRC_PATH not in sys.path:
    sys.path.append(SRC_PATH)

y_train = np.load(
    "data/processed/y_train_ml.npy"
)

class_counts = y_train.sum(axis=0)

print(class_counts)

#policzenie wag dla klasy pozytywnej
total = len(y_train)

pos_weights = total / (2 * class_counts)

print(pos_weights)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

pos_weights_tensor = torch.tensor(
    pos_weights,
    dtype=torch.float32
).to(device)

print(pos_weights_tensor)

```
