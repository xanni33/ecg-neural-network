```python
#setup
import os
import sys
import numpy as np
import matplotlib.pyplot as plt

import torch

PROJECT_PATH = r"C:\Users\cp\PycharmProjects\ECGmodel"

os.chdir(PROJECT_PATH)

SRC_PATH = os.path.join(PROJECT_PATH, "src")

if SRC_PATH not in sys.path:
    sys.path.append(SRC_PATH)

from model import ECGCNN
from interpretability import compute_saliency_map

device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print(device)

#load model
model = ECGCNN(num_classes=5)

model.load_state_dict(
    torch.load(
        "models/ecg_cnn.pth",
        map_location=device
    )
)

model = model.to(device)

model.eval()

print("model loaded")

X_test = np.load(
    "data/processed/X_test_ml.npy"
)

y_test = np.load(
    "data/processed/y_test_ml.npy"
)

print(X_test.shape)
print(y_test.shape)

#wybor pojedynczego przykladu
sample_idx = 0

signal = X_test[sample_idx]

target = y_test[sample_idx]

print(target)

signal_tensor = torch.tensor(
    signal,
    dtype=torch.float32
).unsqueeze(0).to(device)

class_names = [
    "NORM",
    "MI",
    "STTC",
    "CD",
    "HYP"
]

with torch.no_grad():

    logits = model(signal_tensor)

    probs = torch.sigmoid(logits)

probs = probs.cpu().numpy()[0]

for cls, prob in zip(class_names, probs):

    print(f"{cls}: {prob:.3f}")

target_class = np.argmax(probs)

print(
    "explaining:",
    class_names[target_class]
)

saliency = compute_saliency_map(
    model,
    signal_tensor,
    target_class
)

print(saliency.shape)

signal_np = signal[0]

plt.figure(figsize=(16,5))

plt.plot(
    signal_np,
    label="ECG signal"
)

plt.plot(
    saliency,
    alpha=0.7,
    label="Saliency"
)

plt.title(
    f"Saliency Map - {class_names[target_class]}"
)

plt.xlabel("Samples")
plt.ylabel("Amplitude")

plt.legend()

plt.show()

#heatmap overlay
plt.figure(figsize=(16,5))

plt.plot(signal_np)

plt.scatter(
    np.arange(len(signal_np)),
    signal_np,
    c=saliency,
    cmap="hot",
    s=8
)

plt.colorbar(label="Importance")

plt.title(
    f"ECG Interpretability - {class_names[target_class]}"
)

plt.show()

#zautomatyzowanie interpretacji
def explain_sample(sample_idx):

    signal = X_test[sample_idx]
    target = y_test[sample_idx]

    signal_tensor = torch.tensor(
        signal,
        dtype=torch.float32
    ).unsqueeze(0).to(device)

    with torch.no_grad():
        logits = model(signal_tensor)
        probs = torch.sigmoid(logits)

    probs = probs.cpu().numpy()[0]

    pred_class = np.argmax(probs)

    saliency = compute_saliency_map(
        model,
        signal_tensor,
        pred_class
    )

    signal_np = signal[0]

    print("\nTrue label:", target)
    print("Predicted class:", class_names[pred_class])

    for cls, prob in zip(class_names, probs):
        print(f"{cls}: {prob:.3f}")

    # plot 1: signal + saliency
    plt.figure(figsize=(16,5))
    plt.plot(signal_np, label="ECG signal")
    plt.plot(saliency, alpha=0.7, label="Saliency")
    plt.title(f"Saliency Map - {class_names[pred_class]}")
    plt.legend()
    plt.show()

    # plot 2: heatmap
    plt.figure(figsize=(16,5))
    plt.plot(signal_np)
    plt.scatter(
        np.arange(len(signal_np)),
        signal_np,
        c=saliency,
        cmap="hot",
        s=8
    )
    plt.colorbar(label="Importance")
    plt.title(f"ECG Interpretability - {class_names[pred_class]}")
    plt.show()

#losowe probki z roznych klas
np.random.seed(42)

random_indices = np.random.choice(len(X_test), 5, replace=False)

for idx in random_indices:
    explain_sample(idx)

mi_indices = np.where(y_test[:, 1] == 1)[0]

for idx in mi_indices[:3]:
    explain_sample(idx)

class_to_index = {
    "NORM": 0,
    "MI": 1,
    "STTC": 2,
    "CD": 3,
    "HYP": 4
}

def find_misclassified(n=5):

    wrong = []

    for i in range(len(X_test)):

        signal = X_test[i]

        signal_tensor = torch.tensor(signal, dtype=torch.float32).unsqueeze(0).to(device)

        with torch.no_grad():
            logits = model(signal_tensor)
            probs = torch.sigmoid(logits)

        pred = np.argmax(probs.cpu().numpy()[0])
        true = np.argmax(y_test[i])

        if pred != true:
            wrong.append(i)

        if len(wrong) >= n:
            break

    return wrong


wrong_samples = find_misclassified(5)

for idx in wrong_samples:
    print("\n--- MISCLASSIFIED ---")
    explain_sample(idx)

avg_saliency = []

for i in range(50):
    signal = X_test[i]

    signal_tensor = torch.tensor(signal, dtype=torch.float32).unsqueeze(0).to(device)

    with torch.no_grad():
        logits = model(signal_tensor)
        probs = torch.sigmoid(logits)

    pred_class = np.argmax(probs.cpu().numpy()[0])

    saliency = compute_saliency_map(model, signal_tensor, pred_class)

    avg_saliency.append(np.mean(saliency))

print("Mean saliency intensity:", np.mean(avg_saliency))
```
