```python
# ==============================
# imports
# ==============================

import os
import sys
import numpy as np

import torch

from torch.utils.data import DataLoader

from sklearn.metrics import (
    classification_report,
    f1_score
)

PROJECT_PATH = r"C:\Users\cp\PycharmProjects\ECGmodel"

os.chdir(PROJECT_PATH)

SRC_PATH = os.path.join(PROJECT_PATH, "src")

if SRC_PATH not in sys.path:
    sys.path.append(SRC_PATH)

from ecg_dataset import ECGDataset
from model import ECGCNN


# ==============================
# device
# ==============================

device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print(device)


# ==============================
# load dataset
# ==============================

test_dataset = ECGDataset(
    "data/processed/X_test_ml.npy",
    "data/processed/y_test_ml.npy"
)

test_loader = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle=False
)


# ==============================
# load model
# ==============================

model = ECGCNN(num_classes=5)

model.load_state_dict(
    torch.load(
        "models/ecg_cnn.pth",
        map_location=device
    )
)

model = model.to(device)

model.eval()

print("model loaded")


# ==============================
# predictions
# ==============================

all_probs = []
all_targets = []

with torch.no_grad():

    for X_batch, y_batch in test_loader:

        X_batch = X_batch.to(device)

        outputs = model(X_batch)

        probs = torch.sigmoid(outputs)

        all_probs.append(probs.cpu())
        all_targets.append(y_batch)


# tensor -> numpy

y_prob = torch.cat(all_probs).numpy()

y_true = torch.cat(all_targets).numpy()

print(y_prob.shape)
print(y_true.shape)

# ==============================
# threshold tuning
# ==============================

class_names = [
    "NORM",
    "MI",
    "STTC",
    "CD",
    "HYP"
]

best_thresholds = {}

for i, cls in enumerate(class_names):

    best_f1 = 0
    best_thresh = 0.5

    for thresh in np.arange(0.1, 0.9, 0.05):

        preds = (y_prob[:, i] > thresh).astype(int)

        f1 = f1_score(
            y_true[:, i],
            preds,
            zero_division=0
        )

        if f1 > best_f1:
            best_f1 = f1
            best_thresh = thresh

    best_thresholds[cls] = (
        best_thresh,
        best_f1
    )

print(best_thresholds)

thresholds = np.array([
    best_thresholds["NORM"][0],
    best_thresholds["MI"][0],
    best_thresholds["STTC"][0],
    best_thresholds["CD"][0],
    best_thresholds["HYP"][0]
])

y_pred_tuned = (
    y_prob > thresholds
).astype(int)

print(
    classification_report(
        y_true,
        y_pred_tuned,
        target_names=class_names,
        zero_division=0
    )
)
```
