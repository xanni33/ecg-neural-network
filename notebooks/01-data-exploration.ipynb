```python
#00. setup sciezek i importy
import sys
import os
import numpy as np
import pandas as pd
import wfdb

PROJECT_PATH = r"C:\Users\cp\PycharmProjects\ECGmodel"
os.chdir(PROJECT_PATH)
print("current working directory: ", os.getcwd())

SRC_PATH = os.path.join(PROJECT_PATH, "src")
if SRC_PATH not in sys.path:
    sys.path.append(SRC_PATH)

from preprocessing import preprocess_pipeline
from dataset import prepare_dataset
from eda import plot_signal, plot_signal_length_distribution, plot_label_histogram

RAW_PATH = os.path.join(PROJECT_PATH, "data", "raw", "ptbxl")
PROCESSED_PATH = os.path.join(PROJECT_PATH, "data", "processed")
os.makedirs(PROCESSED_PATH, exist_ok=True)



#01. metadane
df = pd.read_csv(os.path.join(RAW_PATH, "ptbxl_database.csv"))
statements = pd.read_csv(os.path.join(RAW_PATH, "scp_statements.csv"), index_col=0)

print("number of records:", len(df))
display(df.head())



#02. budowa poprawnej sciezki
def get_record_path(filename_lr: str) -> str:
    """
    filename_lr: records100/00000/00001_lr
    """
    return os.path.join(RAW_PATH, filename_lr)



#03. poprawiona!! funkcja load_record
def load_record(full_path_no_ext: str):
    #wczytywanie tylko sygnalu
    record = wfdb.rdrecord(full_path_no_ext)
    return record.p_signal[:, 0]



#04. eda
example = df.iloc[0]
record_path = get_record_path(example["filename_lr"])

if os.path.exists(record_path + ".hea"):
    sig = load_record(record_path)
    plot_signal(sig, title=f"example ECG - {example['filename_lr']}")
else:
    print("missing:", record_path)



#05. dystrybucja dlugosci
all_signals = []
#[]

for _, row in df.iloc[:20].iterrows():
    path = get_record_path(row["filename_lr"])
    if os.path.exists(path + ".hea"):
        all_signals.append(load_record(path))

plot_signal_length_distribution(all_signals)



#06. histogram etykiet
all_labels = [
    list(eval(row["scp_codes"]).keys())
    for _, row in df.iloc[:20].iterrows()
]

plot_label_histogram(all_labels)



#07. preprocessing
signals = []
labels = []

for _, row in df.iloc[:20].iterrows():
    path = get_record_path(row["filename_lr"])
    if not os.path.exists(path + ".hea"):
        continue

    sig = load_record(path)
    lbls = list(eval(row["scp_codes"]).keys())

    signals.append(sig)
    labels.append(lbls)


all_segments = []
all_segment_labels = []

for sig, lbls in zip(signals, labels):
    segments = preprocess_pipeline(sig)
    for seg in segments:
        all_segments.append(seg)
        all_segment_labels.append(lbls)

all_segments = np.array(all_segments, dtype=np.float32)
all_segment_labels = np.array(all_segment_labels, dtype=object)

print("segments:", all_segments.shape)
print("labels:", all_segment_labels.shape)



#08. dataset split (zestaw kontrolny)
prepare_dataset(
    all_segments,
    all_segment_labels,
    test_size=0.2
)

#to-do w kolejnych krokach:
#zwiekszenie liczby rekordow
#zmiana etykiety na klasy rytmow lub multi-label scp
#reshape do (N, 1, L) dla modeli 1d-cnn






#potrzebne wykresy do overleafa:
import matplotlib.pyplot as plt
from preprocessing import bandpass_filter, normalize

# wybierz sygnał
signal = signals[0]

filtered = bandpass_filter(signal)
normalized = normalize(filtered)

plt.figure(figsize=(15,5))

plt.plot(signal, label="Original signal", alpha=0.6)
plt.plot(filtered, label="After Butterworth filter", alpha=0.8)

plt.title("ECG signal - before and after filtering")
plt.legend()
plt.show()

#---------------------
from preprocessing import preprocess_pipeline

signal = signals[0]
segments = preprocess_pipeline(signal)

plt.figure(figsize=(15,5))

plt.plot(signal, label="Original signal", alpha=0.5)

for i, seg in enumerate(segments[:5]):  # pokazujemy pierwsze 5 segmentów
    start = i * 720
    end = start + 720
    plt.axvspan(start, end, alpha=0.2)

plt.title("ECG segmentation (2-second windows)")
plt.legend()
plt.show()

def plot_example(df, idx):
    row = df.iloc[idx]
    sig = load_record(get_record_path(row["filename_lr"]))
    plt.plot(sig)
    plt.title(str(eval(row["scp_codes"]).keys()))
    plt.show()

plot_example(df, 0)

label_counts = [len(lbl) for lbl in labels]

plt.hist(label_counts, bins=10)
plt.title("Number of labels per ECG record")
plt.show()

import matplotlib.pyplot as plt
import numpy as np

if 'X' not in globals():
    if 'all_segments' in globals():
        arr = np.array(all_segments, dtype=np.float32)
        # jeśli mamy kształt (N, L) -> zamieniamy na (N, 1, L)
        if arr.ndim == 2:
            X = arr.reshape((-1, 1, arr.shape[1]))
        else:
            X = arr
    elif 'signals' in globals():
        # zbuduj X z pierwszego segmentu każdego sygnału
        built = []
        for s in signals:
            segs = preprocess_pipeline(s)
            if len(segs) > 0:
                built.append(segs[0])
        X = np.array(built, dtype=np.float32)
        if X.ndim == 2:
            X = X.reshape((-1, 1, X.shape[1]))
    else:
        raise RuntimeError("Nie znaleziono `all_segments` ani `signals` — nie można zbudować X")
signal = X[0][0]

#gaussian noise
noise = signal + np.random.normal(0, 0.05, len(signal))

#scaling
scaled = signal * 1.1

#shift
shifted = np.roll(signal, 20)

plt.figure(figsize=(15,10))

plt.subplot(4,1,1)
plt.plot(signal)
plt.title("Original ECG")

plt.subplot(4,1,2)
plt.plot(noise)
plt.title("Gaussian noise")

plt.subplot(4,1,3)
plt.plot(scaled)
plt.title("Amplitude scaling")

plt.subplot(4,1,4)
plt.plot(shifted)
plt.title("Time shifting")

plt.tight_layout()

plt.show()
```
