```python
import matplotlib.pyplot as plt

# liczba rekordow
dataset_sizes = [500, 1000, 5000, 10000]

# twoje wyniki macro f1
f1_scores = [0.266, 0.329, 0.422, 0.55]

plt.figure(figsize=(8,5))

plt.plot(
    dataset_sizes,
    f1_scores,
    marker="o"
)

plt.xlabel("Liczba rekordów treningowych")
plt.ylabel("Macro F1-score")

plt.title("Wpływ liczby rekordów na jakość modelu")

plt.grid(True)

plt.show()
```
